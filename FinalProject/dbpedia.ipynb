```python
import collections

region_to_authors = collections.defaultdict(set)

phil = data["rdf-schema#label"].tolist()
tree = collections.defaultdict(dict)

for i, row in data.iterrows():
    parent = row["rdf-schema#label"]
    infBy = row["influencedBy_label"]
    if not pd.isnull(infBy):
        newInf = infBy.replace("{", "").replace("}", "").split('|')
        print "list of influences for {} is {}".format(parent, newInf)
        
        if (pd.isnull(parent)):
            continue
            
        tree[parent] = collections.defaultdict(dict)
        for c in newInf:
            if c in phil:
                birth_region = "Other"

                i = data[data["rdf-schema#label"]==c]
            
                child =  i["rdf-schema#label"].values
                birth = i["birthPlace_label"].values[0]
                
                if not pd.isnull(birth):                
                    birth = birth.replace("{", "").replace("}", "").split('|')
                    for b in birth:
                        if b in region_by_child:
                            print child, b
                            birth_region = region_by_child[b]                        

                
                print "FOUND", parent, child, "birth:", birth_region
                
                region_to_authors[birth_region].add(child[0])
                
                if birth_region not in tree[parent]:
                    tree[parent][birth_region] = collections.defaultdict(dict)
                tree[parent][birth_region][child[0]] = 1
```

```python
out = []
# print tree

def add_tree(out, tree):
    for i, v in tree.items():
        item = {"name":i, "children":[]}

#         print "==",i, v.keys()
        for s in v:
#             print "region", (s), v[s]
            c_child = []                        
            for c in v[s]:
                if c in tree:
                    print "child",c,"influenced by", tree[c]
                    newtree = {c:tree[c]}
                    add_tree(c_child, newtree)
                else:
                    c_child.append({"name":c})                 

            item["children"].append({"name":s, "children":c_child})
        out.append(item)

add_tree(out, tree)

# print tree["Gottfried Wilhelm Leibniz"]
```
